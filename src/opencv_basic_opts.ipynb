```python
import numpy as np
import cv2
from PIL import Image

# Load an color image
img = cv2.imread('../img/horse.jpeg')

print(type(img))
print(img.size)
```

```python
Image.fromarray(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
# It should be noticed that the order of bands of opencv and pillow images are different.
# It should be transformed as bellow.
# 需要注意的是,opencv存储RGB图像的波段顺序和pillow不同，需要以以下方式转换才能正常显示.
Image.fromarray(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
