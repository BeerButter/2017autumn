Open an image use Pillow.  
用Pillow工具打开一张图片.

```python
# Make sure this program is started under src directory, otherwise the image will not be successfully loaded.
# 需确定程序是在src目录下运行,否则会无法找到图片文件.
im_PIL = Image.open("../img/horse.jpeg")
```

Output some information about an image.  
输出图片信息.

```python
print(im_PIL.format, im_PIL.size, im_PIL.mode)
```

Show an image.  
显示图片.

```python
# show image with system's default image viewer
# 用系统默认的图像查看软件查看图像
im_PIL.show()
```

```python
# a much more easier way to show an image in jupyter-notebook
# jupyter-notebook 下的一种更简单的图片可视化方式
im_PIL
```

Get array reference of an image's data. Print some information of this image.  
获取图像的数据数组的引用,并打印一些信息.

```python
im_np = np.array(im_PIL)
print('type of im_np:', type(im_np))
print('dtype of im_op:', im_np.dtype)
print('shape of im_op (hight,width,bands): ', im_np.shape)
(height,width,bands) = im_np.shape
```

Get the transform of the old image's numpy array.  
创建原图像的转置图像的数据数组.

```python
newwidth = height; newheight = width
im_t_np = np.zeros((newheight,newwidth,bands),dtype=np.uint8)
for ih in range(newheight):
    for iw in range(newwidth):
        for ib in range(bands):
            im_t_np[ih][iw][ib] = im_np[iw][ih][ib]
```

Construct PIL.Image from numpy.ndarray.  
从numpy数组构建PIL.Image.

```python
im_t_PIL = Image.fromarray(im_t_np)
type(im_t_np)
```

Get the first band's gray image of the origin image and construct a new image.  
获取原图像第一波段的灰度图像,并将其构建为一张新的图片.

```python
im_gray_np = np.zeros((height,width),dtype=np.uint8)
for ih in range(height):
    for iw in range(width):
        im_gray_np[ih][iw] = im_np[ih][iw][0]
im_gray_PIL = Image.fromarray(im_gray_np)
im_gray_PIL.show()
```

Save the images.  
存储图片.

```python
im_gray_PIL.save('im_gray.png')
im_t_PIL.save('im_t.png')
```
